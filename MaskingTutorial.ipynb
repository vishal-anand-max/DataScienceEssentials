```python
# ============================
# CELL 2 — CODE
# ============================

import numpy as np
np.set_printoptions(suppress=True)
```

# ============================
# CELL 3 — MARKDOWN
# ============================

"""
### Masking Based on Values

A **value‑based mask** is created by applying a condition to array values.
The result is a boolean array of the same shape.
"""


```python
# ============================
# CELL 4 — CODE
# ============================

A = np.array([1.2, 4.0, 7.5, 10.0, 13.0])

mask = A > 5
masked_A = A[mask]

print("A:", A)
print("Mask:", mask)
print("A[mask]:", masked_A)
```

```python
# ============================
# CELL 5 — CODE
# ============================

# PRACTICE 1:
# Select only even numbers from 0 to 19

B = np.arange(20)
mask = B % 2 == 0
print(B[mask])
```

```python
# PRACTICE 2:
# Mask out negative values

C = np.array([3.5, -2.1, 0.0, 8.8, -1.0])
mask = C >= 0
print(C[mask])
```

# ============================
# CELL 6 — MARKDOWN
# ============================

"""
Boolean indexing **selects elements**, not positions.
The output is always **1‑D**, even if the input is 2‑D.
"""


```python
# ============================
# CELL 7 — CODE
# ============================

A = np.arange(1, 10).reshape(3, 3)
mask = A % 2 == 0

print("A:\n", A)
print("Mask:\n", mask)
print("A[mask]:", A[mask])
```

```python
# ============================
# CELL 8 — CODE
# ============================

# PRACTICE:
# Predict the shape before running

D = np.arange(16).reshape(4, 4)
mask = D > 5
print(D[mask])
```

# ============================
# CELL 9 — MARKDOWN
# ============================

"""
If you want to **preserve shape**, use an **integer mask** (0/1)
and apply it via multiplication.
"""


```python
# ============================
# CELL 10 — CODE
# ============================

A = np.arange(1, 10).reshape(3, 3)
mask_bool = A % 2 == 0
mask_int = mask_bool.astype(int)

print("Integer mask:\n", mask_int)
print("A * mask:\n", A * mask_int)
```

```python
# ============================
# CELL 11 — CODE
# ============================

# PRACTICE:
# Zero out odd numbers but preserve shape

E = np.arange(1, 17).reshape(4, 4)
mask = (E % 2 == 0).astype(int)
print(E * mask)
```

# ============================
# CELL 12 — MARKDOWN
# ============================

"""
Index‑based masks depend on (i, j), not values.

We generate index grids using np.indices.
"""


```python
# ============================
# CELL 13 — CODE
# ============================

A = np.arange(1, 10).reshape(3, 3)
I, J = np.indices(A.shape)

mask_lower = I >= J
mask_upper = I < J

print("Lower‑triangular mask:\n", mask_lower.astype(int))
print("Upper‑triangular mask:\n", mask_upper.astype(int))
```

```python
# ============================
# CELL 14 — CODE
# ============================

# PRACTICE 1:
# Mask where i + j is even

mask = (I + J) % 2 == 0
print(mask.astype(int))
# PRACTICE 2:
# Mask only the diagonal

mask = I == J
print(mask.astype(int))
```

# ============================
# CELL 15 — MARKDOWN
# ============================

"""
NumPy provides direct triangular mask generators.
"""



```python
# ============================
# CELL 16 — CODE
# ============================

A = np.arange(1, 10).reshape(3, 3)

mask_lower = np.tri(*A.shape, dtype=int)
mask_upper = 1 - mask_lower

print("Lower:\n", A * mask_lower)
print("Upper:\n", A * mask_upper)
```

# ============================
# CELL 18 — MARKDOWN
# ============================

"""
np.where(condition, value_if_true, value_if_false)
"""


```python
# ============================
# CELL 19 — CODE
# ============================

A = np.arange(-5, 5)
result = np.where(A >= 0, A, 0)
print(result)
```

```python
# ============================
# CELL 20 — CODE
# ============================

# PRACTICE:
# Replace odd numbers with -1

B = np.arange(10)
result = np.where(B % 2 == 0, B, -1)
print(result)
```

# ============================
# CELL 21 — MARKDOWN
# ============================

"""
### Final Challenge

Create a mask that:
• Keeps only the lower triangle
• AND keeps only even values
• Preserves shape
"""


```python
# ============================
# CELL 22 — CODE
# ============================

A = np.arange(1, 26).reshape(5, 5)
I, J = np.indices(A.shape)

mask = (I >= J) & (A % 2 == 0)
result = A * mask.astype(int)

print(result)
```
